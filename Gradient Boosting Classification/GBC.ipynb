# Gradient Boosting Classification

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Read the train dataset

```python
import pandas as pd
df=pd.read_csv("train_titanic.csv")
df.head()
```

```python
df.info()
```

```python
m=df.isna().sum()
m
```

```python
m[m>0]
```

### Duplicates in dataframe

```python
df.duplicated().sum()
```

### Seperate X and Y (Survived)

```python
X=df.drop(columns=['PassengerId','Name','Ticket','Cabin','Survived'])
Y=df[['Survived']]
```

```python
Y.head()
```

```python
X.head()
```

### Seperate cat and con features

```python
cat=list(X.columns[X.dtypes=='object'])
con=list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

### Create a preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe =Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scaler',StandardScaler())])
```

```python
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('ohe',OneHotEncoder(handle_unknown='ignore'))])
```

```python
pre=ColumnTransformer([('num',num_pipe,con),('cat',cat_pipe,cat)])
```

```python
X_pre=pre.fit_transform(X)
X_pre
```

```python
cols=pre.get_feature_names_out()
cols
```

```python
X_pre=pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

```python
Y.value_counts()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.2,random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.value_counts()
```

```python
ytest.value_counts()
```

### Create a gradient boosting model

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier(learning_rate=0.001,n_estimators=1,max_depth=2,min_samples_split=5,min_samples_leaf=5)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

### Hyperparameter tuning with randomised search CV

```python
params={'learning_rate':[0.01,0.05,0,1,0,2],'n_estimators':[10,50,100,200],'max_depth':[2,3,4,5,6,7],'min_samples_split':[5,6,7,8,9,10]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
gbc=GradientBoostingClassifier()
rscv=RandomizedSearchCV(gbc,param_distributions=params,cv=5,scoring='f1_macro')
rscv.fit(xtrain,ytrain)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

```python
best_gbc=rscv.best_estimator_
best_gbc
```

### Evaluate model performance

```python
best_gbc.score(xtrain,ytrain)
```

```python
best_gbc.score(xtest,ytest)
```

### Predict train and test results

```python
ypred_tr=best_gbc.predict(xtrain)
ypred_ts=best_gbc.predict(xtest)
```

```python
ypred_tr[0:5]
```

```python
ytrain.head()
```

```python
ypred_ts[0:5]
```

```python
ytest.head()
```

### Check the confusion matrix

```python
c=best_gbc.classes_
c
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cf=confusion_matrix(ytest,ypred_ts)
cfd=ConfusionMatrixDisplay(cf,display_labels=c)
cfd.plot()
```

### Check Classification report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_ts))
```
