### Read the store_data.csv

```python
with open('store_data.csv','r') as file:
    lines=file.readlines()
```

```python
lines
```

```python
lines[0:5]
```

```python
a=lines[0].strip() # strip ignore the \n that is new line
a
```

```python
a.split(',')
```

```python
record=[]
for i in lines:
    a=i.strip()
    cart=a.split(',')
    record.append(cart)
```

```python
record
```

```python
record[0:5]
```

###  Data Extraction completed

```python
#support
sup=(7*5)/len(record)
sup
```

```python
# confidence
conf=0.2
```

```python
# lift
lift=3
```

```python
from apyori import apriori
assiociation_rule=apriori(record,min_support=sup,min_confidence=conf,min_lift=3)
assiociation_results=list(assiociation_rule)
```

```python
assiociation_results
```

```python
assiociation_results[0][0]
```

```python
freq_bought=[]
for i in assiociation_results:
    freq_bought.append(i[0])
```

```python
freq_bought
```

### Saving associatio results

```python
with open('Results.txt','w') as File2:
    for i in assiociation_results:
        File2.write(str(i)+'\n')
```
