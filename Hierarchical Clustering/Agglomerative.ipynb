```python
df.columns=['CustomerID', 'Gender', 'Age', 'Income',
       'Spend']
```

```python
df.head()
```

### Seperate X data

```python
X=df[['Age','Income','Spend']]
X.head()
```

### Preprocess the X data

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),
                         ('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre=num_pipe.fit_transform(X)
X_pre.head()
```

```python
X_pre.mean()
```

```python
X_pre.std()
```

### Calculate optimal number of cluster on basis of silhoutee score

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
n=list(range(2,15))
n
```

```python
scores=[]
for i in n:
    #create the model
    model=AgglomerativeClustering(n_clusters=i)
    model.fit(X_pre)
    #predict the model clusters
    lb=model.labels_
    # calculate silhoutte scores
    sil=silhouette_score(X_pre,lb)
    scores.append(sil)
```

```python
scores
```

### Visualize silhoutee scores

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(n,scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhoutee Scores')
plt.title('Silhoutee Scores for various clusters')
plt.grid()
plt.show()
```

### based on above graph for 6 clusters we have highets silhoutee score of 0.42

### Create a agglomerative model with n_cluster=6


```python
model=AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
model.labels_
```

### Save above results in X

```python
X['Label']=model.labels_
```

```python
X[X['Label']==0]
```

```python
for i in range(6):
    print(f'Cluster : {i}\n')
    display(X[X['Label']==i])
    print("=======================================\n")
```

### Centroids

```python
centroids=X.groupby(by='Label').mean()
centroids
```
